```python
def my_product_two_dim_with_threshold(a,b):
    return a[0]*b[0]+a[1]*b[1]+a[2]*b[2]

def get_my_data():
    my_data_x = []
    my_data_x.append([1,0,0])
    my_data_x.append([1,0,1])
    my_data_x.append([1,1,0])
    my_data_x.append([1,1,1])
    my_data_x
    
    my_data_y = []
    my_data_y.append(0)
    my_data_y.append(1)
    my_data_y.append(1)
    my_data_y.append(0)
    my_data_y
    
    return my_data_x,my_data_y
```
